```python
# libs
import stable_baselines3
from stable_baselines3 import DQN
from stable_baselines3.common.evaluation import evaluate_policy
import numpy as np
import random
import gymnasium as gym
```

```python
# ambiente 1
env_dqn = gym.make("LunarLander-v3", continuous=False, gravity=-10.0, enable_wind=False, turbulence_power=1.5)
```

```python
# modelo 1
modelo_dqn = DQN(
    policy="MlpPolicy",
    env=env_dqn,
    learning_rate=1e-4,
    buffer_size=1000000,
    learning_starts=10000,
    batch_size=128,
    gamma=0.99,
    exploration_fraction=0.2,
    exploration_final_eps=0.05,
    verbose=1
)
```

```python
# aprendizado
modelo_dqn.learn(total_timesteps=1000000)
```

```python
resultado_dqn = evaluate_policy(
    modelo_dqn,
    env_dqn,
    n_eval_episodes=100
)

print("DQN - Recompensa média:", resultado_dqn[0])
print("DQN - Desvio padrão:", resultado_dqn[1])
```

```python
obs, info = env_dqn.reset()

for i in range(20):
    action, _states = modelo_dqn.predict(obs, deterministic=True)
    obs, reward, terminated, truncated, info = env_dqn.step(action)

    print(f"(DQN) -> Passo {i+1} | Ação: {action} | Recompensa: {reward}")

    if terminated or truncated:
        obs, info = env_dqn.reset()
```

```python
# modelo 2, alterei o número de neurônios por camada oculta, o padrão é (64, 64)
# alterei também a taxa de aprendizado e o batch_size para valores maiores
modelo_dqn_2 = DQN(
    "MlpPolicy",
    env_dqn,
    learning_rate=5e-4,
    buffer_size=1000000,
    learning_starts=10000,
    batch_size=256,
    gamma=0.99,
    policy_kwargs=dict(net_arch=[256,256]),
    verbose=1
)
```

```python
# treinamento
modelo_dqn_2.learn(total_timesteps=1000000)
```

```python
resultado_dqn_2 = evaluate_policy(
    modelo_dqn_2,
    env_dqn,
    n_eval_episodes=100
)

print("DQN - Recompensa média:", resultado_dqn_2[0])
print("DQN - Desvio padrão:", resultado_dqn_2[1])
```

```python
obs, info = env_dqn.reset()

for i in range(50):
    action, _states = modelo_dqn_2.predict(obs, deterministic=True)
    obs, reward, terminated, truncated, info = env_dqn.step(action)

    print(f"(DQN) -> Passo {i+1} | Ação: {action} | Recompensa: {reward}")

    if terminated or truncated:
        obs, info = env_dqn.reset()
```
